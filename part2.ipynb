1. What are the WHO/CDC BMI categories and their associated health risks? List each category with its BMI range.

Your answer: Underweight: < 18.5 (risk of nutritional deficiency, weakened immune system); Normal weight: 18.5–24.9 (lowest general health risk); Overweight: 25.0–29.9 (increased risk of heart disease, Type 2 diabetes); Obesity Class I: 30.0–34.9 (high risk); Obesity Class II: 35.0–39.9 (very high risk); Obesity Class III: ≥ 40 (extremely high risk)

2. What is insulin resistance, and how does it relate to Type 2 diabetes? Why might insulin levels be an important predictor?

Your answer: Insulin resistance is when the body cells fail to respond properly to insulin, causing glucose to build up in the bloodstream. Over time, this can lead to Type 2 diabetes. Insulin levels might be an important predictor because high insulin can signal the body is compensating for resistance before blood sugar becomes severely elevated.

3. What is the triceps skinfold thickness measurement used for? How does it relate to body composition and diabetes risk?

Your answer: This measurement is used for assessing subcutaneous fat, estimating total body fat percentage, and evaluating nutritional status. Higher triceps skinfold thickness indicates higher body fat percentage, which is associated with increased risk of insulin resistance and Type 2 diabetes. 

4. Why is this dataset specifically from women of Pima Indian heritage? What makes this population unique for diabetes research?

Your answer: The Pima Indian population has one of the highest rates of Type 2 diabetes in the world. Therefore, studying this group helps researchers understand genetic and environmental risk factors, making it valuable for diabetes research.
