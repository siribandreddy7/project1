```python
import pandas as pd

df = pd.read_csv("diabetes.csv")

avg_age = df.groupby("Outcome")["Age"].mean()

print(avg_age)
```

Diabetic patients in this dataset are older on average by about 6 years. This makes clinical sense, because as people get older insulin resistance tends to increase and individuals are more imposed to getting Type 2 diabetes.
